```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el dataset
# Supongamos que el dataset se llama 'diabetes.csv'
data = pd.read_csv('diabetes.csv')

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(data['Glucose'], data['Outcome'], alpha=0.5)

# Agregar etiquetas y título
plt.title('Relación entre Glucosa y Riesgo de Diabetes')
plt.xlabel('Glucosa')
plt.ylabel('Riesgo de Diabetes')

# Mostrar la visualización
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar el estilo de los gráficos de Seaborn
sns.set_style("whitegrid")

# Crear un histograma con diferentes colores para representar las categorías de resultado
plt.figure(figsize=(10, 6))
sns.histplot(data, x='Glucose', hue='Outcome', kde=True, palette='husl', alpha=0.7)

# Agregar etiquetas y título
plt.title('Distribución de Glucosa por Resultado de Diabetes')
plt.xlabel('Glucosa')
plt.ylabel('Frecuencia')

# Mostrar la visualización
plt.legend(title='Resultado')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar el estilo de los gráficos de Seaborn
sns.set_style("whitegrid")

# Crear un grid de diagramas de dispersión
g = sns.FacetGrid(data, col='Outcome', hue='Outcome', height=6)
g.map(plt.scatter, 'Glucose', 'BMI', alpha=0.5)

# Agregar leyenda
g.add_legend(title='Resultado')

# Agregar título
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Relación entre Glucosa, BMI y Riesgo de Diabetes')

# Mostrar la visualización
plt.show()
```
