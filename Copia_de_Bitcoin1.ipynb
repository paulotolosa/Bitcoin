```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/paulotolosa/Bitcoin/main/BTC-USD.csv'

# Lee el archivo CSV en un DataFrame de pandas
df = pd.read_csv(url)

# Convierte la columna de fechas a formato de fecha
df['Date'] = pd.to_datetime(df['Date'])

# Extrae el año de cada fecha
df['year'] = df['Date'].dt.year

# Calcula el promedio de la columna "Volume" para cada año
promedio_volume_por_ano = df.groupby('year')['Volume'].mean()

# Ordena los datos por año
promedio_volume_por_ano = promedio_volume_por_ano.sort_index()

# Crea un gráfico de pastel con el promedio de la columna "Volume" para cada año
plt.figure(figsize=(8, 8))
plt.pie(promedio_volume_por_ano, labels=promedio_volume_por_ano.index, autopct='%1.1f%%', startangle=140)
plt.title('Promedio de la columna "Volume" por año')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python


# Convierte la columna de fechas a formato de fecha
df['Date'] = pd.to_datetime(df['Date'])

# Filtra los datos para incluir solo el año 2015
df_2015 = df[df['Date'].dt.year == 2015]

# Agrupa los datos filtrados por mes y calcula la suma del volumen para cada mes
sum_volume_by_month_2015 = df_2015.groupby(df_2015['Date'].dt.month)['Volume'].sum()

# Crea el histograma
plt.figure(figsize=(10, 6))
sum_volume_by_month_2015.plot(kind='bar', color='skyblue')
plt.title('Suma total de volumen por mes - Año 2015')
plt.xlabel('Mes')
plt.ylabel('Suma total de volumen')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
