```python
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que tienes un DataFrame llamado df que contiene tus datos
# Asegúrate de que la columna 'Date' esté en formato de fecha (puedes convertirla usando pd.to_datetime si no lo está)

# Convertir la columna 'Date' a formato de fecha si no lo está
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar los datos por año
datos_por_anio = df.groupby(df['Date'].dt.year)

# Calcular el máximo valor de Low por año
maximo_low_por_anio = datos_por_anio['Low'].max()

# Calcular el máximo valor de High por año
maximo_high_por_anio = datos_por_anio['High'].max()

# Crear el histograma
plt.figure(figsize=(10, 6))

# Graficar el máximo valor de Low con una barra roja
plt.bar(maximo_low_por_anio.index, maximo_low_por_anio.values, color='red', label='Máximo valor de Low')

# Graficar el máximo valor de High con una barra azul
plt.bar(maximo_high_por_anio.index, maximo_high_por_anio.values, color='blue', label='Máximo valor de High')

# Personalizar el gráfico
plt.title('Máximo valor de Low y High de Bitcoin por año')
plt.xlabel('Año')
plt.ylabel('Precio máximo')
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
